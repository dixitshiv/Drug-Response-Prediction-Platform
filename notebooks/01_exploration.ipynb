```python
# Create mapping from SangerModelID to ModelID
sanger_to_model = model_info[['SangerModelID', 'ModelID']].dropna()
sanger_to_model = sanger_to_model.set_index('SangerModelID')['ModelID'].to_dict()

print(f"Mapping created: {len(sanger_to_model)} Sanger IDs mapped to ModelIDs")

# Get unique Sanger IDs from drug response
drug_sanger_ids = set(drug_response['SANGER_MODEL_ID'].unique())
print(f"\nDrug response has {len(drug_sanger_ids)} unique cell lines")

# Get ModelIDs from expression
expression_model_ids = set(expression_clean.index)
print(f"Expression has {len(expression_model_ids)} unique cell lines")

# Find cell lines present in BOTH datasets
common_cell_lines = []
for sanger_id in drug_sanger_ids:
    if sanger_id in sanger_to_model:
        model_id = sanger_to_model[sanger_id]
        if model_id in expression_model_ids:
            common_cell_lines.append((sanger_id, model_id))

print(f"\nCell lines in BOTH datasets: {len(common_cell_lines)}")
print("\nSample matches (SangerID -> ModelID):")
for sanger, model in common_cell_lines[:5]:
    print(f"  {sanger} -> {model}")
```
